```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

data = pd.read_csv(r'C:\Users\OJH\OneDrive\SKKU\연구자료\Catalyst AI\New Database\Na_W_Mn_Cat_preprocess_operation.csv')

data
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.1, random_state=42)

train = TabularDataset(data=train_data)
test = TabularDataset(data=test_data.drop(columns=['Y(C2)']))
```

```python
# 이렇게 한 줄만 작성하면 내부에서 알아서 학습해줍니다.
predictor = TabularPredictor(label='Y(C2)', eval_metric='rmse').fit(train)
```

```python
ld_board = predictor.leaderboard(train, silent=True)

ld_board
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# 예측하기
pred_y = predictor.predict(test)

print(r2_score(test_data['Y(C2)'], pred_y))
print(mean_squared_error(test_data['Y(C2)'], pred_y))
```

```python
data_scaled = StandardScaler().fit_transform(data)

data_scaled = pd.DataFrame(data_scaled, columns=data.columns)

train_data, test_data = train_test_split(data_scaled, test_size=0.1, random_state=42)

train = TabularDataset(data=train_data)
test = TabularDataset(data=test_data.drop(columns=['Y(C2)']))

predictor = TabularPredictor(label='Y(C2)', eval_metric='rmse').fit(train)

ld_board = predictor.leaderboard(train, silent=True)

ld_board
```
