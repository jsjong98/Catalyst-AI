```python
selected_data = selected_data[['Temperature','pCH4_per_pO2', 'Contact time', 'Y(C2)']]

from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(selected_data)
```

```python
from sklearn.cluster import KMeans

# Applying K-Means Clustering
k = 3  # Example number of clusters, adjust as needed
kmeans = KMeans(n_clusters=k, random_state=0)
clusters = kmeans.fit_predict(scaled_data)

# Adding the cluster ID to the original data
selected_data['Cluster'] = clusters
```

```python
import matplotlib.pyplot as plt

# Visualizing the clusters (using the first two features)
plt.figure(figsize=(10, 6))
for i in range(k):
    plt.scatter(scaled_data[clusters == i, 0], scaled_data[clusters == i, 1], label =f'Cluster {i+1}')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Clusters of Data')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import numpy as np

# Range of k values to try
k_values = range(2, 10)  # Example: trying k from 2 to 9

silhouette_coefficients = []
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(scaled_data)

    # Silhouette Coefficient
    score = silhouette_score(scaled_data, clusters)
    silhouette_coefficients.append(score)

    # Cohesion (Inertia)
    inertia_values.append(kmeans.inertia_)
```

```python
# Plotting the Silhouette Coefficients
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, silhouette_coefficients, marker='o')
plt.title('Silhouette Coefficients for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Coefficient')

# Plotting the Inertia Values
plt.subplot(1, 2, 2)
plt.plot(k_values, inertia_values, marker='o')
plt.title('Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Cohesion)')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt

# eps 값의 범위 설정
eps_values = np.arange(0.1, 20, 0.5)  # 예: 0.1부터 20까지 0.5 간격으로

# 결과를 저장할 리스트
n_clusters_list = []
n_noise_list = []
silhouette_scores = []

for eps in eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=5)
    dbscan_clusters = dbscan.fit_predict(selected_data)

    # 클러스터와 노이즈 포인트 계산
    n_clusters = len(set(dbscan_clusters)) - (1 if -1 in dbscan_clusters else 0)
    n_noise = list(dbscan_clusters).count(-1)

    n_clusters_list.append(n_clusters)
    n_noise_list.append(n_noise)

    # 실루엣 점수 계산
    if n_clusters > 1:
        filtered_data = selected_data[dbscan_clusters != -1]
        filtered_clusters = dbscan_clusters[dbscan_clusters != -1]
        silhouette_score_val = silhouette_score(filtered_data, filtered_clusters)
        silhouette_scores.append(silhouette_score_val)
    else:
        silhouette_scores.append(None)

# eps 값에 따른 클러스터 수와 실루엣 점수 시각화
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(eps_values, n_clusters_list, marker='o')
plt.title('Cluster number vs eps value')
plt.xlabel('eps')
plt.ylabel('Cluster number')

plt.subplot(1, 2, 2)
plt.plot(eps_values, silhouette_scores, marker='o')
plt.title('Silhouette score vs eps value')
plt.xlabel('eps')
plt.ylabel('Silhouette score')

plt.tight_layout()
plt.show()
```
